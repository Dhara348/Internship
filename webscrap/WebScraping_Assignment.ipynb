## 1.Write a python program to display all the header tags from wikipedia.org.

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
heading=soup.find('h2',class_="mp-h2")
```

```python
heading
```

```python
heading.text
```

```python
headings=[]
for i in soup.find_all('h2',class_="mp-h2"):
    headings.append(i.text)
    
headings
```

```python
headings1=[]
for i in soup.find_all('h1',class_="firstHeading"):
    headings1.append(i.text)
    
headings1
```

## 2. Write a python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and make data frame.

```python
page1=requests.get("https://www.imdb.com/chart/top/?ref_=nv_mv_250")
```

```python
page1
```

```python
soup_imdb=BeautifulSoup(page1.content)
```

```python
name=[]
for i in soup_imdb.find_all('td',class_="titleColumn"):
    name.append(i.text)
    
name
```

```python
rating=[]
for i in soup_imdb.find_all('td',class_="ratingColumn imdbRating"):
    rating.append(i.text)
    
rating
```

```python
release_year=[]
for i in soup_imdb.find_all('span',class_="secondaryInfo"):
    release_year.append(i.text)
    
release_year
```

```python
# Making dataframe 
import pandas as pd
df1=pd.DataFrame({'Movie_Name':name,'Released_year':release_year,'Rating':rating})
df2=df1.iloc[:100]
df2
```
