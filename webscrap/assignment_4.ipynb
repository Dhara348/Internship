```python
rank=[]
name=[]
artist=[]
date=[]
views=[]

rk=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[1]')
for j in rk:
    rank.append(j.text)
namee=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[2]/a')
for j in namee:
    name.append(j.text)
ar=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[3]')
for j in ar:
    artist.append(j.text)
view=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[4]')
for j in view:
    views.append(j.text)
dt=driver.find_elements_by_xpath('//table[@class="wikitable sortable jquery-tablesorter"]/tbody/tr/td[5]')
for j in dt:
    date.append(j.text)
```

## 9. Scrape the details most watched tv series of all time from imdb.com.

You have to find the following details:
- A) Name
- B) Year span
- C) Genre
- D) Run time
- E) Ratings
- F) Votes

```python
driver.get('https://www.imdb.com/list/ls095964455/')
```

```python
name=[]
year=[]
Genre=[]
Runtime=[]
rating=[]
votes=[]

n1=driver.find_elements_by_xpath("//h3[@class='lister-item-header']/a")
for i in n1:
    name.append(i.text)
y1=driver.find_elements_by_xpath("//span[@class='lister-item-year text-muted unbold']")
for i in y1:
    year.append(i.text)
g1=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[5]")
for i in g1:
    Genre.append(i.text)
r1=driver.find_elements_by_xpath("//p[@class='text-muted text-small']/span[3]")
for i in r1:
    Runtime.append(i.text)
rat1=driver.find_elements_by_xpath("//div[@class='ipl-rating-star small']/span[2]")
for i in rat1:
    rating.append(i.text)
v1=driver.find_elements_by_name("nv")
for i in v1:
    votes.append(i.text)
```

```python
print(len(name),len(year),len(Genre),len(Runtime),len(rating),len(votes))
```

```python
TVSeries=pd.DataFrame({'Movie Name':name,
                      'Year':year,
                      'Genre':Genre,
                      'Run Time':Runtime,
                      'Rating':rating,
                      'Votes':votes})
TVSeries
```

## 10. Details of Datasets from UCI machine learning repositories.
- Url = https://archive.ics.uci.edu/
You have to find the following details:
- A) Dataset name
- B) Data type
- C) Task
- D) Attribute type
- E) No of instances

```python
driver.get('https://archive.ics.uci.edu/')
```

```python
view=driver.find_element_by_xpath("/html/body/table[1]/tbody/tr/td[2]/span[2]/a").get_attribute('href')
driver.get(view)
```

```python
dataset_name=[]
dttype=[]
task=[]
attribute_type=[]
No_instances=[]

dn=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[1]/table/tbody/tr/td[2]/p/b/a")
for i in dn:
    dataset_name.append(i.text)
dt=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[2]/p")
for i in dt:
    dttype.append(i.text)
t1=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[3]/p")
for i in t1:
    task.append(i.text)
at=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[4]/p")
for i in at:
    attribute_type.append(i.text)
n1=driver.find_elements_by_xpath("/html/body/table[2]/tbody/tr/td[2]/table[2]/tbody/tr/td[5]/p")
for i in n1:
    No_instances.append(i.text)
```

```python
print(len(dataset_name),len(dttype),len(task),len(attribute_type),len(No_instances))
```

```python
data=pd.DataFrame({'Dataset Name':dataset_name,
                       'Data Type':dttype[1:],
                       'Task':task[1:],
                       'Attribute Type':attribute_type[1:],
                       'No of Instance':No_instances[1:]})
data
```
